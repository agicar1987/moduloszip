### Programa de Opciones Sencillo

Este programa te muestra un menú con tres opciones y luego, según tu elección, te da una respuesta diferente. Es una forma básica de usar "condicionales" (`if`, `elif`, `else`).

```python
print("¡Bienvenido! Elige una de estas opciones:")

# Pedimos al usuario que ingrese un número del 1 al 3
opcion = input("1. Mostrar un saludo\n2. Contar hasta 5\n3. Decir un dato curioso\nIngresa el número de tu elección: ")

# Usamos 'if', 'elif' (else if) y 'else' para ver qué opción eligió el usuario
if opcion == '1':
    print("¡Hola! ¿Cómo estás hoy?")
elif opcion == '2':
    print("Contando: 1, 2, 3, 4, 5!")
elif opcion == '3':
    print("Dato curioso: Las abejas pueden volar más rápido que un coche en una zona urbana.")
else:
    print("¡Ups! Esa opción no es válida. Por favor, elige 1, 2 o 3.")
```

**¿Cómo funciona este código?**

1.  **`print(...)`**: Primero, el programa te muestra un mensaje de bienvenida y te presenta las tres opciones.
2.  **`opcion = input(...)`**: Luego, te pide que escribas el número de la opción que quieres y guarda lo que escribas en una variable llamada `opcion`.
3.  **`if opcion == '1':`**: Esta línea pregunta: "¿La variable `opcion` es igual a '1'?". Si es así, ejecuta el `print` de abajo (el saludo).
4.  **`elif opcion == '2':`**: Si la primera condición (`if`) no se cumple, entonces pregunta: "¿Es `opcion` igual a '2'?". Si lo es, ejecuta ese `print` (la cuenta).
5.  **`elif opcion == '3':`**: Si ninguna de las anteriores se cumplió, verifica si `opcion` es igual a '3'. Si lo es, muestra el dato curioso.
6.  **`else:`**: Si ninguna de las opciones anteriores (1, 2 o 3) fue elegida, entonces el `else` se encarga de decir que la opción no es válida.
